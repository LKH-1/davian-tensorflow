# MNIST Dataset Introduction
Most examples are using MNIST dataset of handwritten digits. It has 55,000 examples for training and 10,000 examples for testing. 
The digits have been size-normalized and centered in a fixed-size image, so each sample is represented as a matrix of size 28x28 with values from 0 to 1.

### Usage
In our examples, we are using TensorFlow input_data.py script to load that dataset. It is quite useful for managing our data, and handle:
Dataset downloading
Loading the entire dataset into numpy array:

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("./mnist/", one_hot=True)

# Load data
x_train = mnist.train.images
y_train = mnist.train.labels
y_test = mnist.test.images
y_test = mnist.test.labels
```
